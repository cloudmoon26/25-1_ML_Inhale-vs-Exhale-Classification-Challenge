```python
# ============================================================
# 0) 필요 라이브러리 설치 및 임포트
# ------------------------------------------------------------
!pip install librosa soundfile --quiet

import os, zipfile, random, numpy as np, pandas as pd
from tqdm import tqdm
import librosa, librosa.display
import torch, torch.nn as nn, torch.optim as optim
from torch.optim.lr_scheduler import CosineAnnealingLR, ReduceLROnPlateau
from torch.cuda.amp import GradScaler, autocast
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, precision_recall_curve
import matplotlib.pyplot as plt

# numpy <-> librosa 호환 패치
np.complex = complex
np.float   = float
# ------------------------------------------------------------

# ============================================================
# 1) 데이터 압축 해제
# ------------------------------------------------------------
if not os.path.isdir('train'):
    with zipfile.ZipFile('/content/train.zip', 'r') as z: z.extractall('train')
if not os.path.isdir('test'):
    with zipfile.ZipFile('/content/test.zip', 'r') as z:  z.extractall('test')

# ============================================================
# 2) CSV 로딩 및 라벨 매핑
# ------------------------------------------------------------
train_df = pd.read_csv('/content/train.csv')   # ID, Target(I/E)
train_df['label'] = train_df['Target'].map({'I': 0, 'E': 1})

# ============================================================
# 3) 하이퍼파라미터
# ------------------------------------------------------------
SR           = 16000   # sampling rate
N_MELS       = 128
HOP_LENGTH   = 256
MAX_LEN      = 64      # time frames
TIME_MASK    = 4      # SpecAugment
FREQ_MASK    = 5
BATCH_SIZE   = 64
EPOCHS       = 50
LR           = 1e-3
WEIGHT_DECAY = 1e-4
PATIENCE     = 5
DEVICE       = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# ============================================================
# 4) 오디오 전처리 & 증강 함수
# ------------------------------------------------------------
def add_noise(y, factor=0.005):
    return y + factor * np.random.randn(len(y))

def time_stretch(y, rate): return librosa.effects.time_stretch(y=y, rate=rate)
def pitch_shift(y, sr, n): return librosa.effects.pitch_shift(y, sr=sr, n_steps=n)

def spec_augment(spec, t_mask=TIME_MASK, f_mask=FREQ_MASK):
    spec = spec.copy()
    # time mask
    t0 = random.randint(0, max(0, spec.shape[1]-t_mask)); spec[:, t0:t0+t_mask] = 0
    # freq mask
    f0 = random.randint(0, max(0, spec.shape[0]-f_mask)); spec[f0:f0+f_mask, :] = 0
    return spec

# ============================================================
# 5) Dataset
# ------------------------------------------------------------
class BreathDataset(Dataset):
    def __init__(self, df, audio_dir, augment=False):
        self.df, self.audio_dir, self.augment = df.reset_index(drop=True), audio_dir, augment

    def __len__(self): return len(self.df)

    def __getitem__(self, idx):
        row   = self.df.iloc[idx]
        ID    = row['ID']
        label = row.get('label')   # test set은 없음
        fname = ID.replace('_I_', '_').replace('_E_', '_')
        if not fname.endswith('.wav'):
            fname += '.wav'

        y, _  = librosa.load(os.path.join(self.audio_dir, fname), sr=SR)

        # ---- waveform-level augment --------------------------------------------------
        if self.augment:
            if random.random() < 0.3: y = add_noise(y, factor=random.uniform(0.002,0.01))
            if random.random() < 0.3:
                y = time_stretch(y, random.uniform(0.9,1.1))
                y = np.pad(y, (0, max(0, SR-len(y))))[:SR]
            if random.random() < 0.3: y = pitch_shift(y, SR, random.randint(-1,1))
        # ------------------------------------------------------------------------------

        S = librosa.feature.melspectrogram(y=y, sr=SR, n_mels=N_MELS, hop_length=HOP_LENGTH)
        S = librosa.power_to_db(S, ref=1.0)
        S = np.pad(S, ((0,0),(0,max(0,MAX_LEN-S.shape[1]))), mode='constant')[:, :MAX_LEN]

        if self.augment: S = spec_augment(S)
        tensor = torch.tensor(S).unsqueeze(0).float()       # (1, n_mels, time)

        return (tensor, torch.tensor(label, dtype=torch.long)) if label is not None else (tensor, ID)

# ============================================================
# 6) DataLoader
# ------------------------------------------------------------
train_idx, val_idx = train_test_split(
    np.arange(len(train_df)), test_size=0.2,
    stratify=train_df['label'].values, random_state=42
)
train_loader = DataLoader(BreathDataset(train_df.iloc[train_idx], 'train', True),
                          batch_size=BATCH_SIZE, shuffle=True,  num_workers=2, pin_memory=True)
val_loader   = DataLoader(BreathDataset(train_df.iloc[val_idx],   'train', False),
                          batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)

# ============================================================
# 7) 모델: ResidualBlock + BreathNet (★ 변경 핵심 ★)
# ------------------------------------------------------------
class ResidualBlock(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.conv1 = nn.Conv2d(in_ch, out_ch, kernel_size=3, padding=1, bias=False)
        self.bn1   = nn.BatchNorm2d(out_ch)
        self.relu  = nn.ReLU(inplace=True)
        self.conv2 = nn.Conv2d(out_ch, out_ch, kernel_size=3, padding=1, bias=False)
        self.bn2   = nn.BatchNorm2d(out_ch)

        self.down = (nn.Identity() if in_ch == out_ch
                     else nn.Conv2d(in_ch, out_ch, kernel_size=1, bias=False))

    def forward(self, x):
        identity = self.down(x)
        out = self.relu(self.bn1(self.conv1(x)))
        out = self.bn2(self.conv2(out))
        return self.relu(out + identity)

class BreathNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.features = nn.Sequential(
            ResidualBlock(1, 16),   nn.MaxPool2d(2),   # (16, 64, 32)
            ResidualBlock(16, 32),  nn.MaxPool2d(2),   # (32, 32, 16)
            ResidualBlock(32, 64),  nn.MaxPool2d(2),   # (64, 16,  8)
            ResidualBlock(64, 128), nn.MaxPool2d(2),   # (128, 8,  4)
        )
        self.pool = nn.AdaptiveAvgPool2d((1,1))
        self.cls  = nn.Sequential(
            nn.Flatten(),
            nn.Linear(128, 64, bias=False),
            nn.BatchNorm1d(64), nn.ReLU(inplace=True),
            nn.Dropout(0.3),
            nn.Linear(64, 2)
        )

    def forward(self, x):
        x = self.features(x)
        x = self.pool(x)
        return self.cls(x)

model = BreathNet().to(DEVICE)

# ============================================================
# 8) Optimizer, Scheduler, AMP
# ------------------------------------------------------------
criterion  = nn.CrossEntropyLoss()
optimizer  = optim.AdamW(model.parameters(), lr=LR, weight_decay=WEIGHT_DECAY)
scheduler  = CosineAnnealingLR(optimizer, T_max=EPOCHS, eta_min=1e-5)
scaler     = GradScaler()

# ============================================================
# 9) Train / Validate Loop (+ Early Stopping)
# ------------------------------------------------------------
def train_one_epoch(model, loader):
    model.train(); tloss=tcorrect=ttotal=0
    for x,y in loader:
        x,y = x.to(DEVICE,non_blocking=True), y.to(DEVICE,non_blocking=True)
        optimizer.zero_grad()
        with autocast():
            out = model(x); loss = criterion(out,y)
        scaler.scale(loss).backward(); scaler.step(optimizer); scaler.update()
        tloss += loss.item()*x.size(0); tcorrect += (out.argmax(1)==y).sum().item(); ttotal+=y.size(0)
    return tloss/ttotal, tcorrect/ttotal

@torch.no_grad()
def validate_one_epoch(model, loader):
    model.eval(); vloss=vcorrect=vtotal=0; probs=targets=[]
    probs, targets = [], []
    for x,y in loader:
        x,y = x.to(DEVICE,non_blocking=True), y.to(DEVICE,non_blocking=True)
        out = model(x); loss = criterion(out,y)
        vloss += loss.item()*x.size(0); vcorrect += (out.argmax(1)==y).sum().item(); vtotal+=y.size(0)
        probs.extend(torch.softmax(out,1)[:,1].cpu()); targets.extend(y.cpu())
    auc = roc_auc_score(targets, probs) if len(set(targets))>1 else 0.0
    return vloss/vtotal, vcorrect/vtotal, auc

best, patience = float('inf'), 0
train_losses, val_losses, train_accs, val_accs, val_aucs = [], [], [], [], []

for epoch in range(EPOCHS):
    tr_loss, tr_acc = train_one_epoch(model, train_loader)
    vl_loss, vl_acc, vl_auc = validate_one_epoch(model, val_loader)
    scheduler.step()

    train_losses.append(tr_loss); val_losses.append(vl_loss)
    train_accs.append(tr_acc);   val_accs.append(vl_acc); val_aucs.append(vl_auc)

    print(f"Epoch {epoch+1:02d} | "
          f"Train {tr_loss:.4f}/{tr_acc:.4f} | "
          f"Val {vl_loss:.4f}/{vl_acc:.4f} | AUC {vl_auc:.4f}")

    if vl_loss < best:
        best, patience = vl_loss, 0
        torch.save(model.state_dict(), 'best_model.pt'); print("  ↳ best model saved")
    else:
        patience += 1
        if patience >= PATIENCE:
            print("  ↳ Early stopping"); break

# ============================================================
# 10) 학습 곡선 시각화
# ------------------------------------------------------------
plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.plot(train_losses,label='Train'); plt.plot(val_losses,label='Val'); plt.title("Loss"); plt.legend()
plt.subplot(1,3,2); plt.plot(train_accs,label='Train');  plt.plot(val_accs,label='Val');  plt.title("Acc");  plt.legend()
plt.subplot(1,3,3); plt.plot(val_aucs,label='Val AUC');  plt.title("AUC"); plt.legend(); plt.show()

# ============================================================
# 11) 테스트 추론 & 제출
# ------------------------------------------------------------
test_df   = pd.read_csv('/content/test.csv')
test_loader = DataLoader(BreathDataset(test_df,'test',False),
                         batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)

model.load_state_dict(torch.load('best_model.pt', map_location=DEVICE)); model.eval()
preds, ids = [], []
with torch.no_grad():
    for x, _ids in test_loader:
        probs = torch.softmax(model(x.to(DEVICE)),1)[:,1].cpu()
        preds.extend((probs>0.5).long().numpy()); ids.extend(_ids)

submission = pd.DataFrame({'ID':ids, 'Target':['E' if p==1 else 'I' for p in preds]})
submission.to_csv('submission.csv', index=False)
print("✅ submission.csv 저장 완료")
```
